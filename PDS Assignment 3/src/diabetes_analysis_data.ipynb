```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the diabetes dataset
diabetes_data = pd.read_csv('/content/data_clean/diabetes_clean_data.csv')
```

```python
# Question a)

# set a seed (to ensure work reproducibility) and take a random sample of 25 observations and
# find the mean Glucose and highest Glucose values of this sample and compare these statistics
# with the population statistics of the same variable. You should use charts for this comparison
```

```python
# Set the seed for reproducibility
np.random.seed(50)
```

```python
sample_size = 25
diabetes_sample = diabetes_data.sample(n=sample_size)
```

```python
# Population statistics
population_mean_glucose = diabetes_data['Glucose'].mean()
population_max_glucose = diabetes_data['Glucose'].max()

# Sample statistics
sample_mean_glucose = diabetes_sample['Glucose'].mean()
sample_max_glucose = diabetes_sample['Glucose'].max()

# Print the statistics
print("Population Mean Glucose:", population_mean_glucose)
print("Population Max Glucose:", population_max_glucose)
print("\nSample Mean Glucose:", sample_mean_glucose)
print("Sample Max Glucose:", sample_max_glucose)
```

```python
# Data for plotting
labels = ['Population Mean', 'Sample Mean', 'Population Max', 'Sample Max']
values = [population_mean_glucose, sample_mean_glucose, population_max_glucose, sample_max_glucose]

# Plotting
plt.bar(labels, values, color=['blue', 'green', 'orange', 'red'])
plt.title('Comparison of Glucose Statistics (Population vs Sample)')
plt.ylabel('Glucose Values')
plt.show()

#Save the chart
plt.savefig('/content/results/MeanMaxComparison.png')
```

```python
# The sample mean glucose (113.92) is lower than the population mean glucose (120.89).
# This suggests that, on average, the glucose levels in the sample are lower than the
# average glucose levels in the entire population.

# The maximum glucose value in the sample (196) is slightly lower than the maximum glucose
# value in the population (199). This indicates that the highest glucose level observed in
# the sample is slightly lower than the highest glucose level in the entire population.
```

```python
# In summary, the sample appears to have a slightly lower average glucose level
# and a slightly lower maximum glucose level compared to the entire population.
```

```python
# Question c)

# Using bootstrap (replace= True), create 500 samples (of 150 observation each) from the
# population and find the average mean, standard deviation and percentile for BloodPressure and
# compare this with these statistics from the population for the same variable. Again, you should
# create charts for this comparison. Report on your findings
```

```python
# Number of bootstrap samples and sample size
num_samples = 500
sample_size = 150

# Initialize arrays to store bootstrap statistics
bootstrap_means = []
bootstrap_std_devs = []
bootstrap_percentiles_50 = []
bootstrap_percentiles_75 = []
bootstrap_percentiles_90 = []


# Bootstrap sampling and calculation of statistics
for _ in range(num_samples):
    bootstrap_sample = diabetes_data['BloodPressure'].sample(n=sample_size, replace=True)
    bootstrap_means.append(bootstrap_sample.mean())
    bootstrap_std_devs.append(bootstrap_sample.std())
    bootstrap_percentiles_50.append(np.percentile(bootstrap_sample, 50))
    bootstrap_percentiles_75.append(np.percentile(bootstrap_sample, 75))
    bootstrap_percentiles_90.append(np.percentile(bootstrap_sample, 90))

# Calculate population statistics for BloodPressure
population_mean_bp = diabetes_data['BloodPressure'].mean()
population_std_dev_bp = diabetes_data['BloodPressure'].std()
population_percentile_50_bp = np.percentile(diabetes_data['BloodPressure'], 50)
population_percentile_75_bp = np.percentile(diabetes_data['BloodPressure'], 75)
population_percentile_90_bp = np.percentile(diabetes_data['BloodPressure'], 90)

# Print the population statistics
print("Population Mean BloodPressure:", population_mean_bp)
print("Population Standard Deviation BloodPressure:", population_std_dev_bp)
print("Population 50th Percentile BloodPressure:", population_percentile_50_bp)
print("Population 75th Percentile BloodPressure:", population_percentile_75_bp)
print("Population 90th Percentile BloodPressure:", population_percentile_90_bp)

# Calculate the average of bootstrap statistics
average_bootstrap_mean = np.mean(bootstrap_means)
average_bootstrap_std_dev = np.mean(bootstrap_std_devs)
average_bootstrap_percentile_50 = np.mean(bootstrap_percentiles_50)
average_bootstrap_percentile_75 = np.mean(bootstrap_percentiles_75)
average_bootstrap_percentile_90 = np.mean(bootstrap_percentiles_90)

# Print the average bootstrap statistics
print("\nAverage Bootstrap Mean BloodPressure:", average_bootstrap_mean)
print("Average Bootstrap Standard Deviation BloodPressure:", average_bootstrap_std_dev)
print("Average Bootstrap 50th Percentile BloodPressure:", average_bootstrap_percentile_50)
print("Average Bootstrap 75th Percentile BloodPressure:", average_bootstrap_percentile_75)
print("Average Bootstrap 90th Percentile BloodPressure:", average_bootstrap_percentile_90)
```

```python
# Data for plotting
labels_bp = ['Population Mean', 'Average Bootstrap Mean', 'Population Std Dev', 'Average Bootstrap Std Dev',
             'Population 50th Percentile', 'Average Bootstrap 50th Percentile',
             'Population 75th Percentile', 'Average Bootstrap 75th Percentile',
             'Population 90th Percentile', 'Average Bootstrap 90th Percentile']

values_bp = [population_mean_bp, average_bootstrap_mean, population_std_dev_bp, average_bootstrap_std_dev,
             population_percentile_50_bp, average_bootstrap_percentile_50,
             population_percentile_75_bp, average_bootstrap_percentile_75,
             population_percentile_90_bp, average_bootstrap_percentile_90]

# Plotting
plt.figure(figsize=(12, 8))
plt.bar(labels_bp, values_bp, color=['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink', 'gray', 'cyan', 'magenta'])
plt.title('Comparison of BloodPressure Statistics (Population vs Bootstrap Samples)')
plt.ylabel('BloodPressure Values')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()

#Save the chart
plt.savefig('/content/results/Mean_SD_Percentiles_comparision.png')
```

```python
# The average bootstrap mean (68.97) is very close to the population mean (69.11),
# indicating that the average BloodPressure values from the bootstrap samples are
# consistent with the population mean.
```

```python
# The average bootstrap standard deviation (19.25) is similar to the population
# standard deviation (19.36). This suggests that the variability of BloodPressure
# values in the bootstrap samples is comparable to the variability in the entire
# population.
```

```python
# The percentiles for BloodPressure in the bootstrap samples closely match the
# corresponding percentiles in the population. This suggests that the distribution
# of BloodPressure values in the sample is representative of the distribution in
# the entire population.
```
