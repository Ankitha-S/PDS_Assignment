```python
import pandas as pd
```

```python
train_data_raw = pd.read_csv('/content/data_raw/train.csv')
```

```python
train_data_raw.head()
```

```python
# dropping the unnamed column from the dataset
train_data_raw.drop(train_data_raw.columns[0], axis=1, inplace=True)
```

```python
train_data_raw.head()
```

a) Look for the missing values in all the columns and either impute them (replace with mean, median, or mode) or drop them. Justify your action for this task.

```python
# Identify missing values in all columns
missing_values = train_data_raw.isna().sum()
```

```python
for column in missing_values.index:
    if missing_values[column] > 0:
        if train_data_raw[column].dtype == 'object':
            # Replace missing values with the mode (most common value), for categorical columns
            mode_value = train_data_raw[column].mode()[0]
            train_data_raw[column].fillna(mode_value, inplace=True)
        else:
            # Replace missing values with mean, for numerical columns
            mean_value = train_data_raw[column].mean()
            train_data_raw[column].fillna(mean_value, inplace=True)
```
